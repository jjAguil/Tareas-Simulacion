#### Ejercicio 

Estudiar el comportamiento de las trayectorias al variar $\mu$ y $\sigma$.

```python
from random import gauss
from math import exp
import matplotlib.pyplot as plt
```

```python
def trayectoria(mu,sigma,p0,t):
    l=[p0]
    pt=p0
    for i in range(t):
        pt=pt*exp(mu + sigma*gauss(0,1))
        l.append(pt)
    return l
```

```python

for i in range(10):
    l=trayectoria(0.001,0.04,10,200)
    plt.plot(l)
plt.show()
```

```python
for i in range(10):
    l=trayectoria(0.01,0.00004,10,200)
    plt.plot(l)
plt.show()
```
