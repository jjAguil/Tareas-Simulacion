### Operaciones elemento por elemento (elementwise)



```python
import numpy as np

# crear arreglos
a = np.array([1, 2, 3, 4])
b = np.array([5, 6, 7, 8])
result = a + b
print(result)  # Output: [6 8 10 12]

# Operaciones matemáticas entre arreglos
c = np.array([10, 20, 30, 40])
result = (a + b) * c
print(result)
```

## Operaciones Booleanas
En Numpy, las operaciones Booleanas (lógicas) nos permiten manipular y filtrar
nuestros arreglos basándonos en proposiciones lógicas, es decir condiciones que
evalúan a `True` o `False`.

Numpy permite que las operaciones booleanas se realicen elemento por elemento,
es decir, podemos comparar entre arreglos, o elementos contra arreglos. Esto
dará como resultado un nuevo arreglo, que nos indica el resultado de la
comparación. Entre los operadores más comunes encontramos:
- Igualdad: `==`
- Desigualdad: `!=`
- Mayor que: `>`
- Mayor o igual que: `>=`
- Menor que: `<`
- Menor o igual que: `<=`
   

```python
"""
ejemplo 1.
Crear dos arreglos y comparar si los elementos en arr1 son mayores a su
correspondiente elemento en arr 2
"""
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([3, 3, 2, 4, 1])

result = arr1 > arr2
print("Ejemplo 1")
print(result)


"""
ejemplo 2.
Checar si los elementos de un arreglo son pares
Si el módulo (restante) de un número entre 2 es cero, el número es par
"""
arr3 = np.array([[1, 2, 3, 4, 5, 6], [2, 3, 4, 5, 6, 7]])
result = arr3 % 2 == 1
print("\nEjemplo 2")
print(arr3)
print(result)
print(arr3[~result])
```

## Máscaras Booleanas
Una máscar Booleana es un arreglo que contiene valores `True` o `False`, tal
como los que obtuvimos en la sección anterior. Estos arreglos pueden servir como
filtros para seleccionar elementos específicos, ya sea del mismo arreglo o de un
arreglo diferente. Para lograrlo, usamos la máscara como método de indexado del
arreglo.


```python
"""
ejemplo: indexado avanzado mediante listas
"""
arr = np.array([[1, 2, 3, 4, 5], [2, 3, 4, 5, 6], [6, 5, 4, 3, 2]])
# podemos obtener las columnas 0 y 4 pasando una lista en el índice correspondiente
print("Indexado mediante listas")
print(f"arreglo original:\n{arr}")
print(f"\nrecuperar columnas 0 y 4:\n{arr[:, [0, 4]]}")
print(f"\nrecuperar columnas 1, 2 y 4:\n{arr[:, [1, 2, 4]]}")
```

```python
"""
ejemplo: indexado avanzado mediante arrays
"""
arr = np.array([[1, 2, 3, 4, 5], [2, 3, 4, 5, 6], [6, 5, 4, 3, 2]])
# podemos obtener las columnas 0 y 4 pasando una lista en el índice correspondiente
print("\nIndexado mediante arrays")
print(f"arreglo original:\n{arr}")
arr_cols = np.array([0, 4])
print(f"\nrecuperar columnas 0 y 4:\n{arr[:, arr_cols]}")
arr_cols = np.array([1, 2, 4])
print(f"\nrecuperar columnas 1, 2 y 4:\n{arr[:, arr_cols]}")
```

```python
"""
ejemplo: uso de máscaras booleanas
Recuperamos únicamente los elementos donde la máscara tiene valor `True`
"""
# crear arreglo y encontrar todo aquellos elementos que sean mayores que 2
arr = np.array([1, 2, 3, 4, 5])
mask = arr > 2

# Usando la máscara, obtener un nuevo arreglo con los valores correspondientes
array_filtrado = arr[mask]
print("\nUso de mascaras booleanas")
print(f"arreglo original:\n{arr}")
print(f"mascara:\n{mask}")
print(f"arreglo filtrado:\n{array_filtrado}")

# asignacion de valores de acuerdo a la mascara
print("\nUso de mascaras booleanas")
print(f"arreglo original:\n{arr}")
arr[mask] = 0  # donde la mascara es True, asigna el valor cero
print(f"arreglo despues de asignacion:\n{arr}")
```

También se pueden combinar varias condiciones booleanas mediante el uso de
operadores lógicos, por ejemplo:
- `&` (and): evalúa a `True` si ambos lados de la condición son `True`, en
cualquier otro caso evalúa a `False`
- `|` (or): evalúa a `False` si ambos lados de la condición son `False`, en
cualquier otro caso evalúa a `True`
- `~` (not): cambia el valor de la condición, si era `False` pasa a ser `True` y
viceversa, de `True` a `False`

```python
"""
ejemplo: uso de operadores
"""
arr = np.arange(11)
mask_1 = arr > 2
mask_2 = arr < 8
mask = mask_1 & mask_2
# mask = (arr > 2) & (arr < 8)

print("Ejemplo de operador &")
print(f"arreglo original:\n{arr}")
print(f"mascara 1:\n{mask_1}")
print(f"mascara 2:\n{mask_2}")
print(f"mascara completa:\n{mask}")


mask = mask_1 | mask_2
print("\n\nEjemplo de operador |")
print(f"arreglo original:\n{arr}")
print(f"mascara 1:\n{mask_1}")
print(f"mascara 2:\n{mask_2}")
print(f"mascara completa:\n{mask}")



mask = arr % 2 == 0
mask_neg = ~mask
print("\n\nEjemplo de operador ~")
print(f"arreglo original:\n{arr}")
print(f"mascara original:\n{mask}")
print(f"mascara negada:\n{mask_neg}")
```

## Masked arrays
En Numpy, los *masked arrays* son una estructura de datos que permite trabajar
con arreglos que contengan valores perdidos o datos inválidos, esto mediante la
incorporación de una máscara booleana, que indica cuáles elementos son
válidos y cuáles no. Esta herramienta es muy importante en Data Science, ya que
en muchos casos contamos con valores perdidos. Estos arreglos nos permiten
continuar trabajando con nuestro dataset ignorando por completo lo que no nos
sirve.

La idea detrás de los *masked arrays* es tener dos arreglos: uno con los datos,
y otro con la máscara. La máscara debe tener la misma *shape* que el arreglo de
datos, pero sus valores deben de ser booleanos (`True` o `False`). El valor
`True` nos indica que es un valor inválido.

Para crear un arreglo enmascarado, se usa la función `np.ma.array()`, que permite
pasar un arreglo de datos y uno de máscara simultáneamente.



```python
"""
crear arreglo de datos y mascara
el cuarto valor (-999) será inválido, por ello, la máscara solo evalúa a True en
ese elemento
"""

datos = np.array([1, 2, 3, -999, 5])
mask = np.array([False, False, False, True, False])

# obtener masked array
masked_data = np.ma.masked_array(datos, mask=mask)
print(masked_data)
```

En el ejemplo anterior, el elemento con valor `-999` se marcó como inválido
mediante el arreglo de máscara (haciendo `True` únicamente ese elemento, y
`False` todos los demás). Cuando imprimimos el arreglo, Numpy omite el valor
`-999` y muestre simplemente los símbolos `--` para denotar la presencia de un
elemento inválido.


Toda vez que se cuenta con un masked array, se pueden llevar a cabo diversas
operaciones sobre el. Básicamente, todos los elementos marcados como invalidos
son omitidos en las operaciones matemáticas, por ejemplo: suma, resta,
multiplicación y división.

como: slicing, indexado y broadcasting.

Aparte de crear masked array de modo manual, existen funciones que pueden
crearlos automáticamente, a partir de una condición lógica, por ejemplo, las
